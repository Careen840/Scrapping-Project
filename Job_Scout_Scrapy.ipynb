Import Libraries and Create Driver

```python
import pandas as pd
import time
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager

```

```python
#Create driver
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
```

```python
page_url = "https://www.jobscout24.ch/en/jobs/Junior%20Marketing%20Manager/"
driver.get(page_url)
```

```python

from selenium.webdriver.common.by import By

driver.find_element(By.XPATH, '//button[text()="Accept"]').click()
```

```python
from selenium.webdriver.common.by import By

job_categories = driver.find_element(By.XPATH, '//*[contains(@class, "job-link-detail") and contains(@class, "job-title")]')
job_categories

```

```python
#element = job_categories.text
#job_categories.get_attribute('href')
```

```python
# Navigate to the URL obtained from the 'href' attribute of 'job_categories'
driver.get(job_categories.get_attribute('href'))
```

```python
# Find job elements by class name
job_elements = driver.find_elements(By.CLASS_NAME, 'job-link-detail')

jobs = []

# Loop through each job element and print its text
for job_element in job_elements:
    job_url = job_element.get_attribute('href')
    job_name= job_element.text
    jobs.append({'job_name': job_name, "url": job_url})
    
#print(job_element.text)
    

# Close the WebDriver
#driver.quit()

```

```python
jobs
```

```python
job_list = []

for job in jobs:
    #go to job page
    driver.get(job['url'])
    job_elements= driver.find_elements(By.CLASS_NAME, 'job-link-detail')
    
    for elem in job_elements:
        job_list.append({'job': job['job_name'], 'job': elem.text})
```

```python
job_list
```

```python


#Job_list_df['job'].value_counts().plot(kind= "bar")
Job_list_df = pd.DataFrame(job_list)
print(Job_list_df)

```
